```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["axes.grid"] = True
```

```python
n = 100_000
exp_samples   = np.random.exponential(scale=1.0, size=n)
uniform_samps = np.random.rand(n)
normal_samps  = np.random.randn(n)
```

```python
plt.figure()
plt.hist(exp_samples, density=True, bins=200, histtype="step", label="exponential (scale=1)")
plt.hist(uniform_samps, density=True, bins=200, histtype="step", label="uniform (0,1)")
plt.hist(normal_samps, density=True, bins=200, histtype="step", label="normal (0,1)")

plt.axis([-2.5, 6.0, 0, 1.2])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0, antialiased=True, alpha=0.9, cmap="viridis")
ax.set_title("Z = X^2 + Y^2")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
data = {
    "HP" : np.random.randint(50, 150, 100),   #generates 100 random integers between 50 and 150 for each stat
    "Attack" : np.random.randint(50, 150, 100),
    "Defense" : np.random.randint(50, 150, 100),
    "Sp. Atk" : np.random.randint(50, 150, 100),
    "Sp. Def" : np.random.randint(50, 150, 100),
    "Speed" : np.random.randint(50, 150, 100),
}

df = pd.DataFrame(data)

pearson_corr = df.corr(method="pearson")
spearman_corr = df.corr(method="spearman")

plt.figure(figsize=(14,5))

plt.subplot(1, 2, 1 )
sns.heatmap(pearson_corr, annot=True, cmap="Blues")
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap="Greens")
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
